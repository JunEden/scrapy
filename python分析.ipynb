```python
import pandas as pd
```

```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('python.csv')
javadf = pd.read_csv('java.csv')
```

```python
df = pd.DataFrame(data=df)
javadf = pd.DataFrame(data=javadf)
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
df1 = df.groupby('experience').agg({'salary':'mean'})
```

```python
javadf1 =javadf.groupby('experience').agg({'salary':'mean'})
```

```python
df1,javadf1,df1.plot(kind='bar',figsize=(10,5),title='Python年資薪資平均'),javadf1.plot(kind='bar',figsize=(10,5),title='Java年資薪資平均')
```

```python
df2 = df.groupby('jobArea').agg({'jobArea':'count'})
```

```python
javadf2 =javadf.groupby('jobArea').agg({'jobArea':'count'})
```

```python
df2,javadf2,df2.plot(kind='bar',figsize=(12,12),title='Python縣市職缺'),javadf2.plot(kind='barh',figsize=(15,15),title='Java縣市職缺')
```

```python
df3 = df.groupby('school').agg({'school':'count'})
```

```python
javadf3 = javadf.groupby('school').agg({'school':'count'})
```

```python
df3,javadf3,df3.plot(kind='bar',figsize=(10,10),title='Python學歷職缺'),javadf3.plot(kind='bar',figsize=(10,10),title='Java學歷職缺')
```
